**Task 1: Conceptual Understanding**

1. What is the difference between "Love" and "love" in NLP?
      In NLP, "Love" and "love" are treated as different words because of the difference in capitalization. The system sees them as separate tokens, where "Love" may be used as a name or at the beginning of a sentence, while "love" usually refers to an emotion. This difference can affect how the model understands the text unless we convert all words to lowercase during preprocessing.

2. What happens if stopwords are not removed?
      If stopwords are not removed, the text becomes long and messy with common words like “is”, “the”, and “and”. This makes it harder for the model to find important words, and it can also make the process slower and less accurate.

3. Provide two real-world scenarios where removing stopwords can be harmful.
      Removing stopwords can sometimes change the meaning of a sentence. For example, in “I do not like this,” removing “not” makes it say “I do like this,” which is wrong. Also, in questions like “What is your name?”, removing words like “what” makes the sentence unclear and confusing.

4. Explain the difference between stemming and lemmatization.
      Stemming and lemmatization both change words to a simpler form. Stemming just cuts the word, so it may give a wrong word (like “running” → “runn”). Lemmatization gives the correct base word (like “running” → “run”). So, stemming is fast but not always correct, while lemmatization is slower but more accurate.



**Task 2:Build Advanced Preprocessing Function**


```python
import re

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (e.g., soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters (optional but useful)
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Remove very short words (≤2), but keep "no" and "not"
    words = text.split()
    filtered_words = [w for w in words if len(w) > 2 or w in ["no", "not"]]

    return " ".join(filtered_words)

text = "WOW!!! This is soooo goooood!!! I have 2 dogs. Visit www.test.com or mail me@test.com"
print(preprocess_text(text))
```

**Task 3: Stress Testing**

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in test_sentences:
    cleaned = preprocess_text(text)
    tokens = cleaned.split()

    print("Original Text :", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 50)
```

**Task 4: Token Analytics**


```python
for text in test_sentences:
    cleaned = preprocess_text(text)
    tokens = cleaned.split()

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = round(sum(len(w) for w in tokens) / total_tokens, 2) if total_tokens > 0 else 0

    print("Original Text :", text)
    print("Cleaned Tokens:", tokens)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", avg_length)
    print("-" * 50)
```

**Task 5: Frequency Analysis**

```python
from collections import Counter

all_tokens = []
for text in test_sentences:
    cleaned = preprocess_text(text)
    tokens = cleaned.split()
    all_tokens.extend(tokens)

token_counts = Counter(all_tokens)

print("Top 10 most frequent words:")
for word, count in token_counts.most_common(10):
    print(f"'{word}': {count}")

print("\nTop 5 least frequent words:")
# Get all unique words, then sort them by count to find the least frequent
# Handle cases where there are fewer than 5 unique words
least_common_words = token_counts.most_common()[-5:]
for word, count in reversed(least_common_words):
    print(f"'{word}': {count}")
```

**Task 6: Build Full Pipeline**


```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        cleaned = preprocess_text(text)
        tokens = cleaned.split()

        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

result = full_pipeline(test_sentences)

print("Tokens:", result["tokens"])
print("Clean Sentences:", result["clean_sentences"])
```

**Task 7: Error Handling**


```python
edge_cases = [
    "",
    "😍🔥🔥",
    "1234567890",
]

for case in edge_cases:
    tokens = preprocess_text(case)
    cleaned_sentence = " ".join(tokens)

    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 50)
```
